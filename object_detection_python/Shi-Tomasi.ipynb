```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
flat_chess = cv2.imread('flat_chessboard.png')
#BGR to RGB
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)
plt.imshow(flat_chess)
plt.show()
```

```python
# LOADING IN GRAY SCALE
gray_flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_flat_chess,cmap= 'gray')
plt.show()
```

```python
real_chess = cv2.imread('real_chessboard.jpg')
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2RGB)
plt.imshow(real_chess)
plt.show()
```

```python
gray_real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_real_chess, cmap='gray')
plt.show()
```

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess,64,0.01,10)
```

```python
corners = np.int32(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(flat_chess, (x,y),3,(255,0,0,), -1)
```

```python
plt.imshow(flat_chess)
plt.show()
```

```python
corners = cv2.goodFeaturesToTrack(gray_real_chess,80,0.01, 10)
```

```python
corners = np.int32(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(real_chess,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(real_chess)
plt.show()
```
