```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = plt.imread('face.jpg')
plt.imshow(img)
plt.show()
```

```python
edges =  cv2.Canny(img, threshold1=127, threshold2=127)
plt.imshow(edges)
plt.show()
```

```python
edges =  cv2.Canny(img, threshold1=0, threshold2=255)
plt.imshow(edges)
plt.show()
```

```python
edges =  cv2.Canny(img, threshold1=255, threshold2=0)
plt.imshow(edges)
plt.show()
```

```python
edges =  cv2.Canny(img, threshold1=127, threshold2=0)
plt.imshow(edges)
plt.show()
```

```python
edges =  cv2.Canny(img, threshold1=0, threshold2=127)
plt.imshow(edges)
plt.show()
```

```python
med_val = np.median(img)
med_val
```

```python
lower = int(max(0, 0.7*med_val))
upper = int(min(0, 1.3*med_val))
```

```python
edges = cv2.Canny(img, threshold1=lower, threshold2= upper+100)
plt.imshow(edges)
plt.show()
```

```python
blur = cv2.blur(img, ksize=(7,7))
edges = cv2.Canny(blur, threshold1 = lower, threshold2 = upper+50)
plt.imshow(edges)
plt.show()
```
