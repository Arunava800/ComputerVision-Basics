```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# Getting a blank image

```python
blank_img = np.zeros(shape=(512, 512, 3), dtype = np.int16)
blank_img.shape
```

```python
plt.imshow(blank_img)
```

# Creating a rectangle

```python
cv2.rectangle(img = blank_img, pt1=(400,400), pt2=(300, 300), color=(0,0, 255), thickness = 10)
plt.imshow(blank_img)
```

# Circles

```python
cv2.circle(img = blank_img, center=(100,100), radius=50, color=(0,255,0), thickness=8)
plt.imshow(blank_img)
```

# Line

```python
cv2.line(img=blank_img, pt1=(0,0), pt2=(512,512), color=(255,0,0), thickness=5)
plt.imshow(blank_img)
```

# Filled Circle

```python
cv2.circle(img=blank_img, center=(200,200), radius=60, color=(201,101,12), thickness=-1)
plt.imshow(blank_img)
```

# Draw a line

```python
cv2.line(img= blank_img, pt1=(0,0), pt2=(512,512), color=(102, 255, 255), thickness=5)
plt.imshow(blank_img)
```

# Drawing Images Polygon and Text

```python

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img = blank_img, text="Hello", org=(10,200), fontFace=font, fontScale=4, color=(255, 255, 255), thickness= 3, lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

# Polygon

```python
blank_img = np.zeros(shape=(512, 512,3), dtype= np.int32)
plt.imshow(blank_img)
vertices = np.array([[100,300], [200,200], [400,300],[200,400]], dtype = np.int32)
vertices.shape
cv2.polylines(blank_img, [vertices], isClosed=True, color=(255, 0,0), thickness=10)
plt.imshow(blank_img)
```
